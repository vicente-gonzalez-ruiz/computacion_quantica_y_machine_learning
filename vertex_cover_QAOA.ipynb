```python
from qiskit_optimization.problems import QuadraticProgram
```

```python
qp = QuadraticProgram()
qp.binary_var('x0')
qp.binary_var('x1')
qp.binary_var('x2')
qp.binary_var('x3')

qp.minimize(linear = {'x0':1 , 'x1':1, 'x2':1, 'x3':1})
qp.linear_constraint(linear = {'x0':1, 'x1':1}, sense =">=", rhs = 1)
qp.linear_constraint(linear = {'x0':1, 'x2':1}, sense =">=", rhs = 1)
qp.linear_constraint(linear = {'x1':1, 'x3':1}, sense =">=", rhs = 1)
qp.linear_constraint(linear = {'x2':1, 'x3':1}, sense =">=", rhs = 1)

print(qp.export_as_lp_string())
```

```python
from qiskit_optimization.algorithms import MinimumEigenOptimizer
from qiskit.algorithms import NumPyMinimumEigensolver
np_solver = NumPyMinimumEigensolver()
np_optimizer = MinimumEigenOptimizer(np_solver)
result = np_optimizer.solve(qp)
print(result)
```

```python
from qiskit import Aer
from qiskit.algorithms import QAOA
from qiskit.algorithms.optimizers import COBYLA
from qiskit.utils import QuantumInstance
quantum_instance = QuantumInstance(Aer.get_backend("aer_simulator"),
    shots = 1024)
qaoa = QAOA(optimizer = COBYLA(),
    quantum_instance=quantum_instance, reps = 1)
qaoa_optimizer = MinimumEigenOptimizer(qaoa)
result = qaoa_optimizer.solve(qp)
print(result)
```

```python
print('Variable order:', [var.name for var in result.variables])
for s in result.samples:
    print(s)
```

```python
qp = QuadraticProgram()
qp.binary_var('x0')
qp.binary_var('x1')
qp.binary_var('x2')
qp.binary_var('x3')

# x0+x1+x2+x3 + M((1-x0)(1-x1) + (1-x0)(1-x2) + (1-x1)(1-x3) + (1-x2)(1-x3)) = x0+x1+x2+x3 -9x0 -9x1 -9x2 -9x3 +5x0x1 + 5x0x2 + 5x1x2 + 5x2x3

qp.minimize(linear = {'x0':1 , 'x1':1, 'x2':1, 'x3':1, 'x0':-9, 'x1':-9, 'x2':-9, 'x3':-9},
           quadratic = {('x0', 'x1'):5, ('x0', 'x2'):5, ('x1', 'x3'):5, ('x2', 'x3'):5})

print(qp.export_as_lp_string())
```

```python
from qiskit_optimization.algorithms import MinimumEigenOptimizer
from qiskit.algorithms import NumPyMinimumEigensolver
np_solver = NumPyMinimumEigensolver()
np_optimizer = MinimumEigenOptimizer(np_solver)
result = np_optimizer.solve(qp)
print(result)
```
